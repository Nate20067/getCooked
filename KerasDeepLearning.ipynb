```python
#Importing the Depenencies
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
#Creating an empty array for trianing data and training label
train_labels = [] #Outputs
train_data = [] #Training data input
```

```python
for i in range(50):
  randomYounger =randint(13,64)
  train_data.append(randomYounger)
  train_labels.append(1) #Ou

  randomOlder = randint(65,110)
  train_data.append(randomOlder)
  train_labels.append(0)


for i in range(1000):
  randomYounger =randint(13,64)
  train_data.append(randomYounger)
  train_labels.append(1) #Output

  randomOlder = randint(65,110)
  train_data.append(randomOlder)
  train_labels.append(0) #Input
```

```python
for i in train_data:
  print(i)
```

```python
for i in train_labels:
  print(i)
```

```python
testing_data = []
for i in range(25):
  randomYounger =randint(13,100)
  testing_data.append(randomYounger)

print(testing_data)
testing_data = np.array(testing_data)
```

```python
#Converting lists to np arrays

train_labels = np.array(train_labels)
train_data = np.array(train_data)
```

```python
#Scales down the data
#Transforms the dat between values from 0 to 1
scaler = MinMaxScaler(feature_range=(0,1))
#reshape method reshapes data into 2D raay with one column
#-1 = as many rows needed, 1 means one column
scale_trainingData = scaler.fit_transform(train_data.reshape(-1,1))
```

```python
for i in scaled_trainingData:
  print(i)
```

```python
from keras import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
from keras.metrics import categorical_crossentropy
```

```python
#Building a Sequiantal model
#Sequential model is linear stack of layers
model = Sequential()
#First hidden layer with input shape of 1
model.add(Dense(16, input_shape=(1,)))
#Second hidden layer
model.add(Dense(32))
#Activation feature for model used relu function
model.add(Activation('relu'))
#Output Layer
model.add(Dense(2))
#Output Activation feature
model.add(Activation('softmax'))
```

```python
model =Sequential([
    #First hidden layer with input shape of 1
    Dense(16, input_shape=(1,), activation='relu'),
    #Second Hidden layer
    Dense(32, activation='relu'),
    #Output Layer
    Dense(2, activation='softmax')
])
```

```python
model.summary()
```

```python
#Compiles the model
#SGD is the optimizer minimizes the loss
#Loss is the error from predicted output and true label
#metrics is analyzing how much accuracy is seen from the model
model.compile(SGD(learning_rate=0.001), loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#Trains the model with the training data and training labels,
#epoch represent how many times the data is passed through the model
#Batch_size is how much data data is given to he model each instance
#Validation_split represents the validation data
#Verbose is how much output is seen when training the model
model.fit(scaled_trainingData, train_labels, epochs=25, batch_size=10, validation_split=0.2,verbose=4)
```

```python
model.predict(testing_data, batch_size=5, verbose=0)
predictions = model.predict(testing_data, batch_size=5, verbose=0)
predicted_data = np.argmax(predictions,axis=1)
print("The age at testing data is", testing_data[3], "years " +
      "and the whether the medicine works", predicted_data[3])
```
