```python
#Loading the sklearn library load iris dataset
from sklearn.datasets import load_iris
#load the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
#Loading Pandas
import pandas as pd
#Loading Numpy
import numpy as np
#Setting a random seed
np.random.seed(0)
```

```python
#Creating a variable to store the iris_data
iris_data = load_iris() 
#Creating the Iris dataFrame with colums of iris feature Names
df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
#Looking at the first 5 rows of the dataset
df.head()
```

```python
#Creating a Test and Training the Data
df["flowerType"] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)
##df["flower type"] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)
##df["isTrained"] = np.random.uniform(0, 1, len(df)) <= .75
df = df.drop(["flowerType", "species"],axis=1)
#Viewing the first 5 rows
df.head()
```

```python
#Creating separate dataFrames for both th training as well test data
train, test = df[df["isTrained"] == True], df[df["isTrained"] ==False]
#Prints the number of Training and testing observation datasets
print("Number of observations in the training data: ", len(train))
print("Number of observations in the testing data: ", len(test))
```

```python
#Create a list of the Iris feature Names 
features = df.columns[:4]
#View the features
features
```

```python
#Converting the flower type names into digits i.e. zeros and 1s
y = pd.factorize(train["flower type"])[0] #Converts the flower type column into numerical values
#Training dataset 
print(train.head())
#Testing dataset
print(test.head())
#Viewing the target
y 
```

```python
#Initializes the Random Forest Classifier 
#n_jobs2 prioritizes the number of jobs to run when model is training, speeds up the training
#random_state controls the randomness of samples being used ensures that the data will be the same everytime the code is ran
clf = RandomForestClassifier(n_jobs=2, random_state=0)
#Trains the Classifier
#Creates the RandomForestClassifier fits the classifier with the features the input and y the target label the output
clf.fit(train[features], y) 
```

```python
#Applies the Classifier that was trained 
clf.predict(test[features])
```

```python
#Shows the predictions of first 12 probilities 
x_test = test[features] #Set variable x_test for test[features]
predicted_features = clf.predict_proba(x_test)[0:12]
predicted_features
```

```python
#x_test contains the featured values 
predicted_classes = clf.predict(x_test) #Predicts the label of the class for the samples 
predicted_species = iris_data.target_names[predicted_classes] # maps the predicted class labels to the target names
#Shows the predicted flower types fror the observations
predicted_species[:5]
```

```python
#Actual flower types for first 5 observations
test["flower type"].head()
```

```python
#Creating a confusion matrix
pd.crosstab(test["flower type"], predicted_species, rownames=["Actual Flower Names"], colnames=["Predicted Flower Types"])
```

```python
#Predictiing the flower Types at the given measurements
predicted_species = iris_data.target_names[clf.predict([[5.0, 3.6, 1.4, 2.0],[5.5 ,2.7, 1.1, 2.2]])] 
print("The preditced flower Types at these values are:", predicted_species)
```
