Linear Regression Model

```python
#Import the dependencies
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Load the given data
```

```python
data = pd.read_csv("Advertising.csv")
```

```python
#Look at the data
```

```python
data.head()
```

```python
#Move the columns that are reductant
```

```python
data.drop(['Unnamed: 0'], axis=1)
```

```python
#Plotting the the data(TV ads and Sales)
```

```python
plt.figure(figsize=(16,8))
#Plotting points on x,y axis
plt.scatter(data["radio"], data["sales"], c="blue")
```

```python
#Generate Linear Approximation of the Data
```

```python
x = data['radio'].values.reshape(-1,1)
y = data['sales'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(x,y)
```

```python
#Visualize the best fit line
```

```python
predictions = reg.predict(x)
#Size of the graph
plt.figure(figsize=(16,8))

#PLotting the data
plt.scatter(data["radio"], data["sales"], c="blue")

#plotting the best fit line
plt.plot(data["radio"], predictions, c="red", linewidth=1)

#Labels for graph on x,y axis
plt.xlabel("Money spent on the radio Ads")
plt.ylabel("Sales Made($)")

#Showing the plotted graph
plt.show()
```
